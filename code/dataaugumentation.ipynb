```python
import os
from PIL import Image
from torchvision import transforms
from tqdm import tqdm
import random

# 🔥 SET YOUR DATA PATH
DATA_DIR = r"D:\majorprojectdataset\ISL_CSLRT_Corpus\ISL_CSLRT_Corpus\Frames_Word_Level"

# 🔥 TARGET IMAGES PER CLASS
TARGET_COUNT = 100

# 🔥 AUGMENTATION PIPELINE
augment = transforms.Compose([
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(20),
    transforms.RandomResizedCrop(224, scale=(0.7,1.0)),
    transforms.ColorJitter(brightness=0.4, contrast=0.4, saturation=0.4),
    transforms.RandomAffine(degrees=15, translate=(0.1,0.1)),
])

# 🔥 PROCESS EACH CLASS
for cls in os.listdir(DATA_DIR):
    cls_path = os.path.join(DATA_DIR, cls)

    if not os.path.isdir(cls_path):
        continue

    images = [f for f in os.listdir(cls_path)
              if f.lower().endswith((".jpg",".png",".jpeg"))]

    current_count = len(images)

    if current_count >= TARGET_COUNT:
        print(f"✅ {cls} already has {current_count} images")
        continue

    print(f"⚡ Augmenting {cls}: {current_count} → {TARGET_COUNT}")

    i = 0
    while current_count < TARGET_COUNT:
        img_name = random.choice(images)
        img_path = os.path.join(cls_path, img_name)

        try:
            img = Image.open(img_path).convert("RGB")

            # Apply augmentation
            aug_img = augment(img)

            # Save new image
            new_name = f"aug_{i}_{img_name}"
            aug_img.save(os.path.join(cls_path, new_name))

            current_count += 1
            i += 1

        except:
            continue
```
